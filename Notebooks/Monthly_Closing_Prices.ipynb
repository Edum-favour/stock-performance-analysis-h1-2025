```python
# Import libraries
import yfinance as yf
import pandas as pd
```

```python
# Define tickers
tickers = ["NRG", "PLTR", "HWM", "MOS", "GE", "GEV", "STX", "SMCI", "NFLX", "UBER", "CAH", "CTAS", "IBM", "ED", "FAST", "ABT", "KO", "ES", "ECL", "KVUE", "PM", "CVS", "NEM", "DG", "T", "WELL", "HAS", "HII", "CNP", "EXC", "CL", "F", "HCA", "MAS", "PFE", "O", "VZ"]
```

```python
# Set date range
start_date = "2025-01-01"
end_date = "2025-06-30"
```

```python
# Download monthly data
monthly_prices = []

for ticker in tickers:
    try:
        data = yf.download(ticker, start=start_date, end=end_date, interval="1mo", auto_adjust=True)
        data = data.reset_index()[["Date", "Close"]]
        data["Ticker"] = ticker
        monthly_prices.append(data)
    except Exception as e:
        print(f"Failed to get data for {ticker}: {e}")
    
```

```python
# Combine into one dataframe
df = pd.concat(monthly_prices, ignore_index=True)
print(df.head(37))
```

```python
# Export to excel
df.to_excel("monthly_closing_prices.xlsx")
```
